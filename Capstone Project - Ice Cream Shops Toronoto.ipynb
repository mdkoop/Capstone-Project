```python
#Use the Notebook to build the code to scrape the following Wikipedia page, https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M, in order to obtain the data that is in the table of postal codes and to transform the data into a pandas dataframe.

df1 = pd.read_html('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M',header=0)
df2=df1[0]
df2
```

```python
#Remove Not assigned Borough's and Neighborhoods.  Assign Neighborhood to Borough if Borough is not assigned but Neighborhood is (already done). 

Unassigned = df2[df2['Borough']=='Not assigned'].index
df2.drop(Unassigned,inplace = True)
df2 = df2.reset_index(drop=True)
df2
```

```python
# Display the shape to print the number of rows in the dataframe (df2).

df2.shape
```

```python
#Get the latitude and the longitude coordinates of each neighborhood (assign to df3).

df3 = pd.read_csv('https://cocl.us/Geospatial_data',header=0)
df3 = df3.rename(columns={'Postal Code': 'Postal code'})
df3
```

```python
# Join the dataframes to display latitude and longitude.  This will be named df4

df4 = pd.merge(df2,df3,on = 'Postal code', how = 'left')
df4
```

```python
#Get only Toronto Data (df5)

df5 = df4[df4.Borough.str.contains('Toronto')]
df5 = df5.reset_index(drop=True)
df5
```

```python
def getNearbyVenues(names, latitudes, longitudes, radius=500):
    
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
        print(name)
            
        # create the API request URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            LIMIT)
            
        # make the GET request
        results = requests.get(url).json()["response"]['groups'][0]['items']
        
        # return only relevant information for each nearby venue
        venues_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Neighborhood', 
                  'Neighborhood Latitude', 
                  'Neighborhood Longitude', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(nearby_venues)
```
